# Calibrate Camer and Undistort Images

In this notebook I will be showing you through the steps needed to get calibration data, possibly store them for later retrieval and finally how to use them to undistort images.

This development will go into the first step of my software pipeline intended to remove distortion from camera images and videos for better estimation of relative positioning and lane curvature.

## Libraries and constants needed

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
import glob
import re
import pickle
%matplotlib inline

# Adding a constant from the documentation to help readability:
#   cv2.imread('image.jpg', 0) VERSUS
#   cv2.imread('image.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
if False == hasattr(cv2, 'CV_LOAD_IMAGE_GRAYSCALE'):
    setattr(cv2, 'CV_LOAD_IMAGE_GRAYSCALE', 0)
```

## The calibrate_camera function

This function will retrieve previously obtained calibration data unless forced not to by setting `recalibrate` to `True`.

In the case previous calibration data was not found, it will use the `glob` package to get the filenames of all the calibration chessboard images, load them one by one and apply `cv2.findChessboardCorners` to them.

I admit being guilty with perfectionism in this case as the code has become lengthier than it should be. This is due to non-essential features such as reviewing the chessboard images in optimal human order (e.g. "calibration1", "2", "3" instead of "calibration10" coming second) as well as insisting on using all the images.

Unfortunately, the images are given in such a way that some of them would not pass, mainly image 1, 4 and 5. The first of which can be solved by using a chessboard corner count tuple of (9, 5) instead of the otherwise needed (9, 6). As to the image 4 and 5, I did alter the images to make them pass with the former tuple.

As you can tell, this along with wanting the claibration process to be faster (using the flag CALIB_CB_FAST_CHECK flag hence disabling the histogram equalization and doing it manually) has contributed to how long the code below has become.

Having said that I wish you enjoy reading my code.

```python
def calibrate_camera(show_corners=False, recalibrate=False):

    if False == recalibrate:
        try:
            calibration_file = open('.\\camera_cal\\calibration_data.p', "rb")
            calibration = pickle.load(calibration_file)

            if "camera_matrix" not in calibration.keys() or \
               "distortion_coefficients" not in calibration.keys():
                raise Exception()

            print("Calibration loaded successfully")
            return calibration["camera_matrix"], calibration["distortion_coefficients"]
        except:
            pass
    
    # Get a list of the names of all calibration-image files in the 'camera_cal'
    # directory
    filenames = glob.glob('.\\camera_cal\\calibration*.jpg')
    # Make sure that the current directory is right and that 20 camera calibration
    # images were found, if not raise an exception
    if (len(filenames) != 20):
        raise Exception("20 calibration images not found")

    print('Found 20 calibration images.');

    # Sorts the filenames according to the sequence number in the filename
    filenames.sort(key=lambda filename: int(re.search('[0-9]+', filename).group(0)))

    # To do actual camera calibration two lists are desperately needed: the
    # object points and the camera points, with the first one being tuples
    # corresponding to chessboard corners in the form of (x_index, y_index,
    # 0) and the camera points being the detected (x, y) position in pixels
    # of that corner in each one of the calibration chessboard images
    objpoints = []
    imgpoints = []
    
    for filename in filenames:
        # Read the image in grayscale (discarding full color information)
        gray = cv2.imread(filename, cv2.CV_LOAD_IMAGE_GRAYSCALE)

        # Extract the sequence number from the calibration image's filename
        sequence = int(re.search('[0-9]+', filename).group(0))

        # Find the chessboard corners while noting that images with sequence
        # numbers 1, 3 and 4 will need a different corner matrix size to be
        # parsed correctly by the function
        if any(np.array([1, 4, 5]) == sequence):
            corners_size = (9, 5)
        else:
            corners_size = (9, 6)
        # Note that I modified three images: 3, 4, and 5 both to enable the
        # algorithm to find the chessboard corners as well as to be able to
        # find them fast without needing to apply histogram equalization and
        # the like. I added the three letter 'mod' before the extension in
        # the filenames of the images I altered/modified.
        retval, corners = cv2.findChessboardCorners(gray, corners_size, None, cv2.CALIB_CB_FAST_CHECK)

        if True == retval:
            print("- Calibration image '{}' parsed correctly".format(filename))

            # 'calibrate_camera_objpoints.ipynb'
            objpoints.append( \
                np.dstack(np.concatenate(( \
                    np.mgrid[:corners_size[1],:corners_size[0]][::-1], \
                    [np.zeros(corners_size[::-1])] \
                ))).reshape((-1,3)).astype(np.float32))
            imgpoints.append(corners)
            
            # Draw detected corners in a plot if visualize is set to TRUE
            if True == show_corners:
                # convert gray chessboard to full color for colored corners
                # to stand out otherwise would be difficult to tell
                img = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)
                # Draw corners on the color-converted img
                cv2.drawChessboardCorners(img, corners_size, corners, True)
                # Add a sub image (aka subplot) in a grid of 5 rows and 4
                # columns with the sequence number of the image as its order
                # in the grid
                plt.subplot(5, 4, sequence)
                # Insert the subimage into the bigger plot with no title and
                # no axes (xticks is set to [] and so is yticks)
                plt.imshow(img)
                plt.xticks([])
                plt.yticks([])
        else:
            print("- Calibration image '{}' could NOT BE PARSED.".format(filename))

    # Show the composed grid of images if visualize was set to TRUE
    if True == show_corners:
        # Title of the entire thing can be set with 'suptitle'
        plt.suptitle("Detected Chessboard Corners", \
                     fontsize = 24)
        # leave only 2% spacing from the left, right and bottom sides. Leave 10%
        # from the top to accommodate the suptitle
        plt.subplots_adjust(left=0.02, bottom=0.02, right=0.98, top=0.90, wspace=0.0, hspace=0.0)
        # Show the composed grid or plot of subplots with a uptitle
        plt.show()

    # Now we have all of our calibration data in objpoints and imgpoints
    retval, matrix, distortion, rvecs, tvecs = \
            cv2.calibrateCamera(objpoints, imgpoints, (1280, 720), None, None)

    if retval:
        print("Calibration succeeded")
        camera_calibration = {}
        camera_calibration["camera_matrix"] = matrix
        camera_calibration["distortion_coefficients"] = distortion

        # Store camera calibration results for later use
        pickle.dump(camera_calibration, open('.\\camera_cal\\calibration_data.p', "wb"))
        
        return matrix, distortion
    else:
        raise Exception("Calibration failed")
```

## Testing the function without visualizing the corners

Here we are testing the function above while forcing it to actually load the files and locate the corners. This is to demonstrate functionality even when calibration data from a previous run is available.

```python
matrix, distortion = calibrate_camera(recalibrate=True)
```

## Testing the function with cached calibration data

Here we call the function without forcing it to over the entire process all again, we store the data in new matrices and compare for verifying the correctness of the storing-retrieving functionality.

```python
matrix_retrieved, distortion_retrieved = calibrate_camera()

print("\n")
print("Camera matrix matching: ", "Yes" if \
      np.array_equal(matrix_retrieved, matrix) else "No")
print("Distortion coefficients matching: ", "Yes" if \
      np.array_equal(distortion_retrieved, distortion) else "No")
```

Last but not least we call the function while asking it to show the detected corners.

```python
%matplotlib inline

calibrate_camera(show_corners=True, recalibrate=True)
```

## Undistort Images

Using the calibration data, mainly the camera matrix and the distortion coefficients we are able to undistort images through `cv2.undistort` which is easy enough to use not to call for a wrapper:

```python
def test_camera_calibration(matrix, distortion):
    # Get filenames of testing images
    filenames = glob.glob(r'.\test_images\test?.jpg')

    # Confirm that we are in the right directory for this
    if len(filenames) != 6:
        raise Exception('Please make sure that the notebook is in the same folder as the test images')

    # Loop over the filenames
    for filename in filenames:
        # Read in the image and undistort it using the calibration
        # data passed to the function
        img = mpimg.imread(filename)
        img_after = cv2.undistort(img, matrix, distortion)

        # Plot the distorted and the undistorted images side by side
        # for comparison
        plt.subplot('121')
        plt.imshow(img)
        plt.title("Distorted")
        plt.xticks([])
        plt.yticks([])
        plt.subplot('122')
        plt.imshow(img_after)
        plt.title("Undistorted")
        plt.xticks([])
        plt.yticks([])

        plt.suptitle(filename, fontsize = 24)
        plt.subplots_adjust(left=0.02, bottom=0.02, right=0.98, top=0.90, wspace=0.0, hspace=0.0)
        plt.show()
        
test_camera_calibration(matrix, distortion)
```
