# Combined Color Gradient Binary Images

So here we are and we have a couple of functions that could help us. I will be mentioning the code I developed in some of the other notebooks with a minor modification: instead of a returning binary images of the dtype 'np.uint8' I will be modifying both calls to return color and gradient binary images of type 'np.bool8': (Please run the code below)

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
import numpy as np
import glob
import re
%matplotlib inline

# 'percentage_binary_thresholding.ipynb'
def percentages_to_thresholds(img_gray, lower_percentage, upper_percentage):
    norm_hist = (np.cumsum(np.histogram(img_gray, 255, (0.0, 255.0))[0]) / \
                 img_gray.size).astype(np.float32)
    values = np.arange(0, 255)[((norm_hist >= lower_percentage) & \
                                (norm_hist <= upper_percentage))]
    
    return np.min(values), np.max(values)

# 'gradient_experiment.ipynb'
def get_binary_gradient(img, low_thresh=0.95, high_thresh=1.0):
    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2HLS_FULL)[:,:,1]
    
    gradient_x = np.absolute(cv2.Sobel(img_gray, cv2.CV_64F, 1, 0))
    gradient_x = 255 * gradient_x / np.max(gradient_x)
    
    binary = np.zeros_like(img_gray).astype(np.bool8)
    low_thresh, high_thresh = percentages_to_thresholds(gradient_x, \
        low_thresh, high_thresh)
    
    binary[(gradient_x >= low_thresh) & (gradient_x <= high_thresh)] = True
    return binary

# 'color_scheme_experiment.ipynb'
def Sally(img):
    S = cv2.cvtColor(img, cv2.COLOR_RGB2HLS_FULL)[:,:,2]
    A = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)[:,:,1]
    
    return np.sqrt( \
                S.astype(np.uint32) * \
                A.astype(np.uint32) \
                  ).astype(np.uint8)

# 'color_scheme_experiment.ipynb'
def get_binary_color(img, low_thresh = 0.9, high_thresh = 1.0):
    img_gray = Sally(img)
    
    low_thresh, high_thresh = percentages_to_thresholds(img_gray, \
        low_thresh, high_thresh)
    
    img_binary = np.zeros_like(img_gray).astype(np.bool8)
    img_binary[(img_gray > low_thresh) & (img_gray < high_thresh)] = True
    
    return img_binary
```

## Testing on Static Images (using the Green-Blue illustration technique)

Here, one of the RGB color channels will designate information donated by color thresholding (using upper 10% of pixels) whereas the other used color channel will designate information donated by the gradient x (i.e. cv2.Sobel) thresholding (this time using the upper 5% of pixels).

If in doubt about what I mean by 10% and 5% of pixels or what the 'percentages_to_thresholds' above does, please refer to the other notebook 'percentage_binary_thresholding.ipynb'.

```python
import glob
import re
%matplotlib inline

filenames = glob.glob('test?.jpg')
if len(filenames) != 6:
    raise Exception('Make sure the test images are in the same folder as the notebook')

for sequence, filename in enumerate(filenames):
    img = mpimg.imread(filename)
    
    binary = np.zeros(img.shape[:2]).astype(np.bool8)
    binary = np.dstack((binary, get_binary_color(img), get_binary_gradient(img)))

    plt.subplot(3, 2, sequence + 1)
    plt.imshow(binary)
    plt.title(filename)
    plt.xticks([])
    plt.yticks([])
    
    bare_filename = re.search(r'([^.\\/]+)\.jpe?g$', filename).group(1)
    mpimg.imsave("{}-CombinedRGB.jpg".format(bare_filename), binary)

plt.show()
print("A total of {} files were saved into disk for study.".format(len(filenames)))
```

## Testing on Video

```python
# Import everything needed to edit/save/watch video clips
from moviepy.editor import VideoFileClip
from IPython.display import HTML
```

```python
# This is the main function in the processing which utilizes both of
# get_binary_color and get_binary_gradient to construct an RGB image
# with one non-utilized channels and two binary channels.
def process_image(img):
    binary_R = 255 * np.zeros(img.shape[:2]).astype(np.uint8)
    binary_G = 255 * get_binary_color(img).astype(np.uint8)
    binary_B = 255 * get_binary_gradient(img).astype(np.uint8)
    binary = np.dstack((binary_R, binary_G, binary_B))
     
    return binary

# The video processing and storing code
input_video = VideoFileClip("project_video.mp4")
output_video = input_video.fl_image(process_image)

output_filename = 'project_video_binary.mp4'
%time output_video.write_videofile(output_filename, audio=False)
```

## Combined Undistort and Binary Thresholding for Lane Detection

Here we call out for out calibration function from 'calibrate_camera.ipynb' and we use that to implement the same functionality as above, however while giving out a black and white binary image and while undistorting the video frames first before thresholding them as before.

```python
# 'calibrate_camera.ipynb'
def calibrate_camera(show_corners=False, recalibrate=False):

    if False == recalibrate:
        try:
            calibration_file = open('.\\camera_cal\\calibration_data.p', "rb")
            calibration = pickle.load(calibration_file)

            if "camera_matrix" not in calibration.keys() or \
               "distortion_coefficients" not in calibration.keys():
                raise Exception()

            print("Calibration loaded successfully")
            return calibration["camera_matrix"], calibration["distortion_coefficients"]
        except:
            pass
    
    # Get a list of the names of all calibration-image files in the 'camera_cal'
    # directory
    filenames = glob.glob('.\\camera_cal\\calibration*.jpg')
    # Make sure that the current directory is right and that 20 camera calibration
    # images were found, if not raise an exception
    if (len(filenames) != 20):
        raise Exception("20 calibration images not found")

    print('Found 20 calibration images.');

    # Sorts the filenames according to the sequence number in the filename
    filenames.sort(key=lambda filename: int(re.search('[0-9]+', filename).group(0)))

    # To do actual camera calibration two lists are desperately needed: the
    # object points and the camera points, with the first one being tuples
    # corresponding to chessboard corners in the form of (x_index, y_index,
    # 0) and the camera points being the detected (x, y) position in pixels
    # of that corner in each one of the calibration chessboard images
    objpoints = []
    imgpoints = []
    
    for filename in filenames:
        # Read the image in grayscale (discarding full color information)
        gray = cv2.imread(filename, cv2.CV_LOAD_IMAGE_GRAYSCALE)

        # Extract the sequence number from the calibration image's filename
        sequence = int(re.search('[0-9]+', filename).group(0))

        # Find the chessboard corners while noting that images with sequence
        # numbers 1, 3 and 4 will need a different corner matrix size to be
        # parsed correctly by the function
        if any(np.array([1, 4, 5]) == sequence):
            corners_size = (9, 5)
        else:
            corners_size = (9, 6)
        # Note that I modified three images: 3, 4, and 5 both to enable the
        # algorithm to find the chessboard corners as well as to be able to
        # find them fast without needing to apply histogram equalization and
        # the like. I added the three letter 'mod' before the extension in
        # the filenames of the images I altered/modified.
        retval, corners = cv2.findChessboardCorners(gray, corners_size, None, cv2.CALIB_CB_FAST_CHECK)

        if True == retval:
            print("- Calibration image '{}' parsed correctly".format(filename))

            # 'calibrate_camera_objpoints.ipynb'
            objpoints.append( \
                np.dstack(np.concatenate(( \
                    np.mgrid[:corners_size[1],:corners_size[0]][::-1], \
                    [np.zeros(corners_size[::-1])] \
                ))).reshape((-1,3)).astype(np.float32))
            imgpoints.append(corners)
            
            # Draw detected corners in a plot if visualize is set to TRUE
            if True == show_corners:
                # convert gray chessboard to full color for colored corners
                # to stand out otherwise would be difficult to tell
                img = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)
                # Draw corners on the color-converted img
                cv2.drawChessboardCorners(img, corners_size, corners, True)
                # Add a sub image (aka subplot) in a grid of 5 rows and 4
                # columns with the sequence number of the image as its order
                # in the grid
                plt.subplot(5, 4, sequence)
                # Insert the subimage into the bigger plot with no title and
                # no axes (xticks is set to [] and so is yticks)
                plt.imshow(img)
                plt.xticks([])
                plt.yticks([])
        else:
            print("- Calibration image '{}' could NOT BE PARSED.".format(filename))

    # Show the composed grid of images if visualize was set to TRUE
    if True == show_corners:
        # Title of the entire thing can be set with 'suptitle'
        plt.suptitle("Detected Chessboard Corners", \
                     fontsize = 24)
        # leave only 2% spacing from the left, right and bottom sides. Leave 10%
        # from the top to accommodate the suptitle
        plt.subplots_adjust(left=0.02, bottom=0.02, right=0.98, top=0.90, wspace=0.0, hspace=0.0)
        # Show the composed grid or plot of subplots with a uptitle
        plt.show()

    # Now we have all of our calibration data in objpoints and imgpoints
    retval, matrix, distortion, rvecs, tvecs = \
            cv2.calibrateCamera(objpoints, imgpoints, (1280, 720), None, None)

    if retval:
        print("Calibration succeeded")
        camera_calibration = {}
        camera_calibration["camera_matrix"] = matrix
        camera_calibration["distortion_coefficients"] = distortion

        # Store camera calibration results for later use
        pickle.dump(camera_calibration, open('.\\camera_cal\\calibration_data.p', "wb"))
        
        return matrix, distortion
    else:
        raise Exception("Calibration failed")

# Get calibration data, highly likely to be retrieved from disk
matrix, distortion = calibrate_camera()

# The video processing routine which simply combines the boolean
# or binary masks from color and gradient and transforms them into
# a black and white frame.
def process_image(img):
    # combining here means joining using the logical 'or' operator
    combined_binary = get_binary_color(img) | get_binary_gradient(img)
    
    img_R = 255 * combined_binary.astype(np.uint8)
    img_G = 255 * combined_binary.astype(np.uint8)
    img_B = 255 * combined_binary.astype(np.uint8)
    img_RGB = np.dstack((img_R, img_G, img_B))
     
    return img_RGB

# The video processing and storing code
input_video = VideoFileClip("project_video.mp4")
output_video = input_video.fl_image(process_image)

output_filename = 'project_video_binary_undistorted.mp4'
%time output_video.write_videofile(output_filename, audio=False)
```
