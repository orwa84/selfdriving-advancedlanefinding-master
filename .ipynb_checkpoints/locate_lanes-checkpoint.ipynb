# Locate Lanes

I will start this notebook by attempting to get screenshots from binary, undistorted video which was the result of the other notebook 'combined_color_gradient.ipynb'. Then I will be using these images to carry preliminary tests to locate the lane lines.

```python
import cv2

video_capture = cv2.VideoCapture('project_video_binary_undistorted.mp4')
if True != video_capture.isOpened():
    raise Exception('Please make sure that this notebook is in the same directory as the test videos')
    
fps = int(video_capture.get(cv2.CAP_PROP_FPS))
success = True
counter = 1
while success:
    for n in range(fps):
        success, img = video_capture.read()
        if not success:
            break
    
    if not success:
        break
    cv2.imwrite('test_binary_frame{}.jpg'.format(counter), img)
    counter += 1
    
print("{} frames saved into the disk for using as test images".format(counter))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for sequence, filename in map((lambda n: (n, 'test_binary_frame{}.jpg'.format(n))), range(1, 5)):
    try:
        f = open(filename)
        f.close()
    except:
        raise Exception('Please make sure that this notebook is in the same directory as the test frames')

    img = plt.imread(filename)
    plt.subplot(2, 2, sequence)
    plt.title(filename, fontsize=8)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    
plt.subplots_adjust(wspace=0.02)
plt.show()
```

## Masking

```python
import matplotlib.image as mpimg
import numpy as np
import glob
import re

# This function was borrowed from the first project
def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

filenames = glob.glob('test_binary_frame*.jpg')
for filename in filenames:
    
    sequence = re.search(r'([0-9]+)\.jpe?g$', filename).group(1)
    
    img = mpimg.imread(filename)
    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    h, w = img_gray.shape[:2]
    a, e, b, c, f, d = [0.35, 0.05, 0.05, 0.55, 0.75, .95]
    img_masked = region_of_interest(img_gray, np.array([[ \
                [e*w, f*h], \
                [a*w, c*h], \
                [(1.0-a)*w, c*h], \
                [(1.0-e)*w, f*h], \
                [(1.0-b)*w, d*h], \
                [b*w, d*h] \
                ]], np.int32))
    
    mpimg.imsave('test_masked_frame{}.jpg'.format(sequence), img_masked, cmap="gray")
    
print("{} frames masked and saved into the disk for inspection".format(len(filenames)))
    
```

## HoughLinesP

```python
import math

def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    mb_lines = []
    for line in lines:
        for x1, y1, x2, y2 in line:
            m = (y2 - y1) / (x2 - x1)
            b = y1 - m * x1
            
            if not math.isinf(m):
                mb_lines.append([m, b])

    mb_lines = np.array(mb_lines, dtype=np.float64)
    
    # now that we have lines in slope-intersection form, we apply
    # our slope-based lane identification criteria
    left_slope_criteria  = (mb_lines[:,0] < -0.4) & (mb_lines[:,0] > -1.2)
    right_slope_criteria = (mb_lines[:,0] >  0.4) & (mb_lines[:,0] <  1.2)
    
    # If failed to recover right or left lane lines, skip
    if 0 == np.count_nonzero(left_slope_criteria) or \
       0 == np.count_nonzero(right_slope_criteria):
        return
        
    lanes = []
    # Left lane: slope 'm' < -0.4
    lanes.append(np.average(mb_lines[left_slope_criteria], 0))
    # Right lane: slope 'm' > 0.4
    lanes.append(np.average(mb_lines[right_slope_criteria], 0))
    
    # Point at which the two lines intersect: (b2 - b1) / (m1 - m2)
    apex_x = int(round((lanes[1][1] - lanes[0][1]) / (lanes[0][0] - lanes[1][0])))
    apex_gap = 15
        
    for m, b in lanes:
        if (m < 0):
            cv2.line(img, \
                     (0, int(round(b))), \
                     (apex_x - apex_gap, int(round(b + (apex_x - apex_gap) * m))), color, thickness)
        if (m > 0):
            cv2.line(img, \
                     (img.shape[1], int(round(b + img.shape[1] * m))), \
                     (apex_x + apex_gap, int(round(b + (apex_x + apex_gap) * m))), color, thickness)
            
filenames = glob.glob('test_masked_frame*.jpg')

for filename in filenames:
    sequence = re.search(r'([0-9]+)\.jpe?g$', filename).group(1)
    
    img_gray = cv2.imread(filename, 0)
    lines = cv2.HoughLinesP(img_gray, 2, np.pi/64, 30, np.array([]), 20, 1000)
    
    img_RGB = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB)
    draw_lines(img_RGB, lines)
    mpimg.imsave('test_straighlanes_frame{}.jpg'.format(sequence), img_RGB)
    
print("{} frames processed and saved into the disk for inspection".format(len(filenames)))
```
