# Experimenting with Color Schemes

In this notebook I want to show a simple yet an effective experiment that I carried to discover which color channels were of potential for detecting lane lines.

-----------------------------
I started the experiment by writing a generator Python functions that fed me the color schemes available in the 'cv2' module, one by one, assuming the image is currently in the 'RGB' color scheme or any other color scheme (called the source_scheme below):

```python
import cv2

def color_schemes(source_scheme='RGB'):
    # Go over all the attributes or names defined within the cv2 library
    for member in dir(cv2):
        # Compose a prefix to help us identify valid color conversions from source_scheme
        prefix = 'COLOR_{}2'.format(source_scheme)
        # Once such a valid color conversion constant is found, return it along with the
        # name of the scheme in textual form (the part that follows the prefix)
        if member.startswith(prefix):
            yield getattr(cv2, member), member[len(prefix):]
            
# Let's test the generator function!
print("Available color schemes in OpenCV:")
for scheme_constant, scheme_name in color_schemes():
    print("- {} (conversion constant from RGB={})".format(scheme_name, scheme_constant))
```

----------------------------
Pretty neat, huh?

If you think that is cool, then let me show you another generator function which I wrote, this one is to obtain the different channels, once by one, from any image array:

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline

# img here is a numpy image array with one or more channels
def channels(img):
    for channel in range(1 if img.ndim < 3 else img.shape[2]):
        if img.ndim < 3:
            # an img with a single channel has a single channel id of '0'
            yield img[:,:], 0
        else:
            # for multiple channels, we return the image of the channel
            # as well as the index for that channel
            yield img[:,:,channel], channel

# Load an RGB color image of the road
img = mpimg.imread('test1.jpg')

# Plot the different channels using our generator function
for channel_img, channel_index in channels(img):
    plt.subplot(1, 3, channel_index + 1)
    plt.imshow(channel_img, cmap="gray")
    plt.title("Channel {}".format(channel_index))
    plt.xticks([])
    plt.yticks([])

plt.show()
    
```

## Time for the actual test

So now that I showed you how to use my generator function tricks let me show you the code I have written to survey all the color channels I could use for my lane line detections.

This code fragment below does just that! It will attempt every possible color conversion and store, as a grayscale image, all the color channels contained in those color conversions. The files are saved into the disk to be reviewed and compared in detail.

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

# Load an RGB color image of the road
img = mpimg.imread('test2.jpg')

# Iterate over all color conversions from RGB
channel_counter = 0
for scheme_value, scheme_name in color_schemes():
    # Convert the image of the road to that color scheme
    img_converted = cv2.cvtColor(img, scheme_value)
    # Iterate over any channels contained in that converted image
    for channel_img, channel_index in channels(img_converted):
        # Store them separately as grayscale images
        filename = "test-{}-{}.jpg".format(scheme_name, channel_index)
        mpimg.imsave(filename, channel_img, cmap="gray")
        # Maybe plot it too to make it easier to see the results here
        channel_counter += 1
        if channel_counter < 4:
            plt.imshow(channel_img, cmap="gray")
            plt.title("{} channel {}".format(scheme_name, channel_index))
            plt.xticks([])
            plt.yticks([])
            plt.show()
            
print("Total of {} different color channels were explored and saved into the disk.". \
        format(channel_counter))
```

## An idea

An idea that I had by looking into these pictures is a way to combine the S channel from the HLS_FULL color scheme and the A channel from the LAB channel.

I will call this combination the Sally color channel.

```python
import glob
import re

# Note that I am assuming that img is in the 'RGB' color scheme which we will go to
# get the Sally hybrid color channel for
def Sally(img):
    S = cv2.cvtColor(img, cv2.COLOR_RGB2HLS_FULL)[:,:,2]
    A = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)[:,:,1]
    
    return np.sqrt( \
                S.astype(np.uint32) * \
                A.astype(np.uint32) \
                  ).astype(np.uint8)

# Get filenames of all the 6 test files in the folder
filenames = glob.glob('test?.jpg')

# Make sure we are in the righ folder
if len(filenames) != 6:
    raise Exception('Make sure the test images are in the same folder as the notebook')

counter = 1
for filename in filenames:
    # Load an RGB test image of the road
    img = mpimg.imread(filename)

    # Get the sally hybrid channel and show it
    img_sally = Sally(img)
    plt.subplot('32' + str(counter))
    plt.imshow(img_sally, cmap="gray")
    plt.title("{} channel Sally".format(filename))
    plt.xticks([])
    plt.yticks([])
    counter += 1
    
    # Maybe also save it for further study
    bare_filename = re.search(r'([^.\\/]+)\.jpe?g$', filename).group(1)
    plt.imsave("{}-sally.jpg".format(bare_filename), img_sally, cmap="gray")
    
plt.show()
 
```

## Setting a binary condition on sally

Last by not least in this experimental notebook I wish to show how can a binary condition be applied to sally to result in a white and a black image which we can then use for steering the self-driving car vehicle or combine with gradient-based information for a more robust lane line detection.

```python
filenames = glob.glob("*-sally.jpg")

if len(filenames) != 6:
    raise Exception('Make sure the test images are in the same folder as the notebook')

counter = 0
for filename in filenames:
    img = cv2.imread(filename)
    img_binary = np.zeros_like(img).astype(np.float32)
    img_binary[(img > 150)] = 1
    
    # Get the sally hybrid channel and show it
    plt.subplot('32' + str(counter))
    plt.imshow(img_binary, cmap="gray")
    plt.title("{} binary".format(filename))
    plt.xticks([])
    plt.yticks([])
    counter += 1
    
    # Maybe also save it for further study
    bare_filename = re.search(r'([^.\\/]+)\.jpe?g$', filename).group(1)
    plt.imsave("{}-binary.jpg".format(bare_filename), img_binary, cmap="gray")
    
plt.show()
```

## get_binary_color()

Here I am condensing the above to a single call, get_binary_color() which uses Sally above. I wish to use that to combine color binary information with gradient-derived binary information to build my pipeline.

```python
import glob
import re


# 'percentage_binary_thresholding.ipynb'
def percentages_to_thresholds(img_gray, lower_percentage, upper_percentage):
    norm_hist = (np.cumsum(np.histogram(img_gray, 255, (0.0, 255.0))[0]) / \
                 img_gray.size).astype(np.float32)
    values = np.arange(0, 255)[((norm_hist >= lower_percentage) & \
                                (norm_hist <= upper_percentage))]
    
    return np.min(values), np.max(values)

# 'color_scheme_experiment.ipynb'
def Sally(img):
    S = cv2.cvtColor(img, cv2.COLOR_RGB2HLS_FULL)[:,:,2]
    A = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)[:,:,1]
    
    return np.sqrt( \
                S.astype(np.uint32) * \
                A.astype(np.uint32) \
                  ).astype(np.uint8)

# 'color_scheme_experiment.ipynb'
def get_binary_color(img, low_thresh = 0.9, high_thresh = 1.0):
    img_gray = Sally(img)
    
    low_thresh, high_thresh = percentages_to_thresholds(img_gray, \
        low_thresh, high_thresh)
    
    img_binary = np.zeros_like(img).astype(np.uint8)
    img_binary[(img_gray > low_thresh) & (img_gray < high_thresh)] = 255
    
    return img_binary
    

filenames = glob.glob("test?.jpg")

if len(filenames) != 6:
    raise Exception('Make sure the test images are in the same folder as the notebook')

counter = 0
for filename in filenames:
    img = cv2.imread(filename)
    img_binary = get_binary_color(img)
    
    # Get the sally hybrid channel and show it
    plt.subplot('32' + str(counter))
    plt.imshow(img_binary, cmap="gray")
    plt.title("{} binary".format(filename))
    plt.xticks([])
    plt.yticks([])
    counter += 1
    
    # Maybe also save it for further study
    bare_filename = re.search(r'([^.\\/]+)\.jpe?g$', filename).group(1)
    plt.imsave("{}-preferred.jpg".format(bare_filename), img_binary, cmap="gray")
    
plt.show()
```
